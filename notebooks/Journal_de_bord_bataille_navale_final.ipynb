## Le projet en bref
Il nous a été donné comme consigne de créer une bataille navale en python, par groupe de 2.
Dans le cadre de ce projet, le cahier des charges qui nous a été donné est le suivant:

### Cahier des charges initial

* Grille : taille peut-elle être changée ? *10x10*
* Bateaux : combien ? tailles ? modifiable ? *2;3;3;4;5*
* Placement des bateaux sur la grille aléatoirement : attention aux intersections
* Utilisateur doit pouvoir tirer
    - définition de la case: attention aux mauvaises saisies
    - feed-back : dans l'eau - touché - coulé
* Contrainte de temps ou de nbr de coups
* Jeu doit s'arrêter - recommencer : menu ?
* Score : nbr de coups et/ou temps
* Affichage avec la tortue
    - grille
    - feed-back
    - infos contrainte et score
* Jeu utilisable sans la tortue
* Notebook : explication et suivi du projet + code optionnel
* Module (fichier python) executable en tant que script : **python batailleNavale.py**

Nous allons maintenant expliquer et présenter la manière dont nous avons résolu chacun des points ci-dessus.

## La grille de jeu
Pour dessiner la grille, nous avons décidé de la décrire à l'aide de la classe `Grille` qui stocke une liste d'instances de la classe `Case` (voir explications pour la classe `Case` ci dessous). La taille de la grille est modifiable facilement dans notre code, ce qui nous a permis d'en faire un paramètre pour l'utilisateur. 
Cette classe possède aussi des méthodes statiques permettant de convertir les coordonnées entre différents systèmes.

### Cases
Chaque case est une instance de la classe `Case`. Elles possèdent chacune une position et un état et peuvent appartenir à un bateau. La position est la position du coin inférieur gauche de la case dans le système de coordonnées de la tortue.

### État
Nous avons une autre classe qui décrit l'état de la case. C'est une valeur entre 0 et 5 dont la signification est décrite par l'énumération `Etat` :
- `Etat.VIDE` décrit les cases blanches et n'appartenant à aucun bateau
- `Etat.BATEAU_INTACT` signifie que la case est blanche mais appartient à un bateau
- `Etat.DANS_L_EAU` veut dire que l'utilisateur a tiré un coup dans l'eau sur la case
- `Etat.TOUCHE` signifie que le joueur a tiré sur la case et qu'un bateau s'y trouvait, mais que ce bateau n'est pas coulé
- `Etat.COULE` décrit les cases appartenant à un bateau coulé

## Bateaux
Nous avons décidé d'avoir 5 bateaux de taille 2,3,3,4 et respectivement 5 cases. (Nous nous sommes inspirés de la bataille navale classique) Nous avons décidé de laisser ce paramètre fixe pour ne pas trop compliquer notre code.

### Modélisation des bateaux
Une classe `AbstractBateau` définit tous les bateaux indépendamment de leur type. Chaque type de bateau est décrit par sa propre classe héritant d'`AbstractBateau`.

## Placement des bateaux
Pour faciliter le placement des bateaux, nous avons décidé de les triller par taille décroissante.
Nous détérminons ensuite la première case du bateau aléatoirement, puis nous cherchons la deuxième case du bateau selon le sens vertical, puis haut/bas ou horizontal puis droite/gauche (choix à nouveau aléatoire).
Nous réitérons ce procédé jusqu'à avoir toutes les cases du bateau. Nous contrôlons si la dernière case fait encore partie de la grille et si toutes les cases sont encore libres. 
Si toutes les conditions sont réunies, alors le bateau est placé. Le cas échéant nous recommençons la boucle.
Nous avons une sécurité en cas d'impossibilité à placer les bateaux: si plus d'une seconde est nécessaire pour placer le bateau, une erreur est affichée. 

## Interaction avec l'utilisateur
L'utilisateur doit entrer au clavier les coordonnées de la case qu'il veut atteindre dans la console. Une fonction "nettoie" ce que l'utilisateur entre et vérifie sa saisie (l'utilisateur peut écrire en majuscule ou en minuscule, cela ne posera pas de problème. En revanche, il doit d'abord placer la lettre puis le chiffre, sinon un message d'erreur s'affiche.
Lorsque l'utilisateur tire, nous colorons la case en fonction de son état; blanc = vide, bleu=dans l'eau, orange = touché, rouge = coulé. De plus, nous affichons l'état de la case en dessous de la grille.

## Contraintes
À l'aide de la fonction `time` du module du même nom il nous a été possible de définir une contrainte de temps qui est affichée en haut à gauche (rafraîchie à chaque nouvelle saisie de case). Nous avons aussi une contrainte de nombre de tirs.

## Fin de partie
Lorsque l'utilisateur a gagné, respectivement perdu, une ligne de texte le lui en informe. Il n'y a pas de score à proprement parler, en revanche le nombre de coups et de secondes restantes reste visible le temps que l'utilisateur quitte, respectivement recommence le jeu. 
Une deuxième ligne de texte indique à l'utilisateur comment accéder au menu.

## Menu
Au début et à la fin de chaque partie un menu est affiché. Pour sélectionner la commande il suffit d'entrer le numéro de la commande au clavier dans la console. 
Notre menu se compose comme suit:
1. Continuer
2. Nouvelle partie
3. Paramètres
4. Quitter

Un message de demande de confirmation est affiché pour les options 2 et 4.
Nos différents paramètres sont les suivants:

### Difficulté
Nous avons décidé d'avoir 3 niveaux de difficulté (facile, moyen, difficile)
Ce paramètre est modifiable dans le menu.
Il définit, selon la taille de la grille, un nombre de coups et un temps maximum (si ces paramètres sont réglés sur "auto").

### Nombre maximum de coups
Ce paramètre est défini automatiquement lors du choix de la difficulté. Néanmoins, l'utilisateur a la possibilité de choisir lui même le nombre maximal de tirs qu'il s'autorise.

### Temps maximum
Ce paramètre est également défini automatiquement lors du choix de la difficulté. Néanmoins, l'utilisateur a aussi la possibilité de choisir lui même une limite temporelle.

### Taille de la grille
L'utilisateur peut entrer la taille du côté de la grille avec laquelle il souhaite jouer. Pour pouvoir placer les bateaux il faut avoir minimum 6 cases de côté. Nous avons fixé le maximum à 26, car il y a 26 lettres dans l'alphabet. Étendre la possibilité à plus aurait impliqué des modifications du code, pour quelque chose qui selon nous n'en vaut pas la peine.
Si l'utilisateur entre un nombre incompatible la question lui est à nouveau posée

## Affichage avec la tortue
Nous avons une classe `Tortue` qui définit les fonctions d'affichage avec la tortue. 

### Vitesse de dessin
Lors de notre première utilisation de la tortue dans le cadre du cours d'informatique, nous avons pu constater la lenteur de la tortue pour effectuer des images complexes, y compris en mettant sa vitesse au maximum. Il existe cependant un moyen efficace de résoudre ce problème, il suffit de désactiver le fraichissement automatique de l'écran en appelant `screen.tracer(0, 0)` sur la tortue. Le désavantage à cela est qu'il faut rafraîchir l'écran manuellement lorsque quelque chose a été dessiné, mais cela rend le rendu considérablement plus rapide.

## Jeu utilisable sans la tortue
Notre jeu est utilisable dans la console sans problème étant donné que nous y affichons aussi la grille et tous les retours!
